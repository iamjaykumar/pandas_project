how='outer': This argument specifies the type of merge to perform. An outer merge includes all rows from both DataFrames, even if there are no matching values in the specified merge key(s). If a key exists in one DataFrame but not the other, the resulting row will have NaN (Not a Number) values for the columns from the DataFrame where the key was missing. This is similar to a SQL FULL OUTER JOIN [

```python
pd.merge(data1, data2)
```

how='cross': This argument specifies the type of merge to perform. A cross merge creates all possible combinations of rows from the two DataFrames. For every row in the first DataFrame (data1), it is paired with every row in the second DataFrame (data2). The resulting DataFrame will have a number of rows equal to the product of the number of rows in data1 and the number of rows in data2. There is no matching of values between the DataFrames; it simply creates a combination of all rows from both sides.

```python
pd.merge(data1, data2, how='cross')
```

* left: use only keys from left frame, similar to a SQL left outer join;
  preserve key order.
* right: use only keys from right frame, similar to a SQL right outer join;
  preserve key order.
* outer: use union of keys from both frames, similar to a SQL full outer
  join; sort keys lexicographically.
* inner: use intersection of keys from both frames, similar to a SQL inner
  join; preserve the order of the left keys.
* cross: creates the cartesian product from both frames, preserves the order
  of the left keys.

The apply() method in pandas is used to apply a function along an axis of a DataFrame or Series. This means you can use it to perform an operation row-wise or column-wise.

Here's a breakdown:

Applying a function: You pass a function to apply(). This function will then be executed on each element (if applied to a Series) or on each row or column (if applied to a DataFrame).
Axis: You specify the axis along which you want to apply the function:
axis=0 (default): Applies the function to each column.
axis=1: Applies the function to each row.

```python
def test(x):
  return x.sum()
df1.apply(test)
```

Here, the apply() method is called on the DataFrame df1. The test function is passed as the argument to apply(). By default, .apply() operates on each column of the DataFrame (equivalent to axis=0). So, for each column in df1, the test function is executed. The test function will calculate the sum of all the values within that column. The result of this operation will be a pandas Series where the index corresponds to the column names of df1, and the values are the sum of each respective column.

.cumsum(): This is a pandas function that calculates the cumulative sum. It works by adding the current element to the sum of all preceding elements in the series. For example, the first value in the resulting series will be the first value in column 'a'. The second value will be the sum of the first and second values in 'a', the third value will be the sum of the first, second, and third values in 'a', and so on.

```python
df4['a'].cumsum() #Cumulative sum
```
